## 3.1
 1. Create a tensor a from `list(range(9))`. Predict then check what the size, offset, and strides are.
 2. Create a new tensor using `b = a.view(3, 3)`. What does view do? Check that a and b share the same storage.
 3. Create a tensor `c = b[1:,1:]`. Predict then check what the size, offset, and strides are.


 1. As `a` is a one-dimension tensor, the size should be `(1,)`, and the stride is `(9)` because we have to skip `9` elements to reach the next one along axis `0`. The offset of the first element is `0`.
 2. The view is a function that creates another tensor, but with a different shape, based on an existing tensor. `b` shares the same storage as `a` since this function is build for avoiding explicit data copy.
 3. The first element of `c` is `4`, which is the fifth element in the storage, thus the offset is `4`. The size should be `(2,2)` and the strides are `(3,1)` according to the definition of `b`.

```python
a = torch.tensor(list(range(9)))
print(f'element a: size = {a.size()}, offset = {a.storage_offset()}, strides = {a.stride()}')

b = a.view((3,3))
print('element b:', 'Same storage?', a.storage().data_ptr() == b.storage().data_ptr())

c = b[1:, 1:]
print(f'element c: size = {c.size()}, offset = {c.storage_offset()}, strides = {c.stride()}')
```

## 4.1 
Take several pictures of red, blue, and green items with your phone or other digital camera. Load each image and convert them to tensors. For each image tensor, use the `.mean()` method to get a sense of how bright the image is. Now take the mean of each channel of your images. Can you identify the red, green, and blue items from only the channel averages?

Yes, since the there is a relationship between color channel and the tensor, i.e. 0 - red, 1 - green, 2 - blue. We can see that the channel with the highest mean value is likely to be the exactly the picture with the object in the according color.

```python
facts = ['red', 'green', 'blue']
files = [f'fig/{f}.jpg' for f in facts]
n = len(files)
fig = plt.figure(figsize=(20, 10))
totensor = ToTensor()
for i in range(n):
    pil = Image.open(files[i])
    img_tensor = totensor(pil)
    brightness = img_tensor.mean()
    channel_mean = img_tensor.mean(axis=[1,2])
    ax = plt.subplot(1, n, i+1)
    ax.title.set_text(f'{facts[i]}: {brightness:.3f}-{channel_mean}')
    ax.imshow(pil)
    plt.axis("off")
plt.show()
```

## 5.1
Redefine the model to be $w_2 t_u^2 + w_1 t_u + b$

1. What parts of the training loop, etc. needed to change to accommodate this redefinition?
  - What parts were agnostic to swapping out the model?
2. Is the resulting loss higher or lower after training?
  - Is the actual result better or worse?
  
  
1. The `model` function and the derivatives function need to be changed since we introduce the weight parameter for $x^2$.
2. Seeing from the result below, the loss is smaller and the result is better.

```python
t_c = [0.5,  14.0, 15.0, 28.0, 11.0,  8.0,  3.0, -4.0,  6.0, 13.0, 21.0]
t_u = [35.7, 55.9, 58.2, 81.9, 56.3, 48.9, 33.9, 21.8, 48.4, 60.4, 68.4]
t_c = torch.tensor(t_c)
t_u = torch.tensor(t_u)
n_samples = t_u.shape[0]
n_val = int(0.2 * n_samples)
shuffled_indices = torch.randperm(n_samples)
train_indices = shuffled_indices[:-n_val]
val_indices = shuffled_indices[-n_val:]
train_t_u = t_u[train_indices]
train_t_c = t_c[train_indices]
val_t_u = t_u[val_indices]
val_t_c = t_c[val_indices]
train_t_un = 0.1 * train_t_u
val_t_un = 0.1 * val_t_u
```

```python
# Original Version

def model(t_u, w, b):
    return w * t_u + b

def loss_fn(t_p, t_c):
    squared_diffs = (t_p - t_c)**2
    return squared_diffs.mean()

def dmodel_dw(t_u, w, b):
    return t_u

def dmodel_db(t_u, w, b):
    return 1.0

def dloss_fn(t_p, t_c):
    dsq_diffs = 2 * (t_p - t_c) / t_p.size(0)
    return dsq_diffs

def grad_fn(t_u, t_c, t_p, w, b):
    dloss_dtp = dloss_fn(t_p, t_c)
    dloss_dw = dloss_dtp * dmodel_dw(t_u, w, b)
    dloss_db = dloss_dtp * dmodel_db(t_u, w, b)
    return torch.stack([dloss_dw.sum(), dloss_db.sum()])

def plot_result(t_p, t_u, t_c):
    fig = plt.figure(figsize=(12, 8))
    plt.xlabel("Fahrenheit")
    plt.ylabel("Celsius")
    plt.plot(t_u.numpy(), t_p.detach().numpy())
    plt.plot(t_u.numpy(), t_c.numpy(), 'o')
    plt.show()

def training_loop(n_epochs, optimizer, params, train_t_u, val_t_u, train_t_c, val_t_c):
    for epoch in range(1, n_epochs + 1):
        train_t_p = model(train_t_u, *params)
        train_loss = loss_fn(train_t_p, train_t_c)

        with torch.no_grad():
            val_t_p = model(val_t_u, *params)
            val_loss = loss_fn(val_t_p, val_t_c)
            assert val_loss.requires_grad == False

        optimizer.zero_grad()
        train_loss.backward()
        optimizer.step()
        training_stat = f'Train: {train_loss:.3f}  Validation: {val_loss:.3f}'
        print('{: <{padding}}'.format(training_stat, padding=75), end='\r', flush=True)

def calc_forward(t_u, t_c, is_train):
    with torch.set_grad_enabled(is_train):
        t_p = model(t_u, *params)
        loss = loss_fn(t_p, t_c)
    return loss
```

```python
# Original Train

w = torch.ones(())
b = torch.zeros(())
w1 = torch.ones(())

params = torch.tensor([1.0, 0.0], requires_grad=True)
learning_rate = 1e-1
optimizer = torch.optim.Adam([params], lr=learning_rate)

training_loop(
    n_epochs = 3000,
    optimizer = optimizer,
    params = params,
    train_t_u = train_t_un,
    val_t_u = val_t_un,
    train_t_c = train_t_c,
    val_t_c = val_t_c)

t_un = 0.1 * t_u
t_p = model(t_un, *params)
plot_result(t_p, t_u, t_c)
```

```python
# Functions after modification

def model(t_u, w, w1, b):
    return w * t_u **2 + w1* t_u + b

def loss_fn(t_p, t_c):
    squared_diffs = (t_p - t_c)**2
    return squared_diffs.mean()

def dmodel_dw(t_u, w, b):
    return t_u ** 2

def dmodel_dw1(t_u, w, b):
    return t_u

def grad_fn(t_u, t_c, t_p, w, b):
    dloss_dtp = dloss_fn(t_p, t_c)
    dloss_dw = dloss_dtp * dmodel_dw(t_u, w, b) 
    dloss_db = dloss_dtp * dmodel_db(t_u, w, b)
    dloss_dw1 = dloss_dtp * dmodel_dw1(t_u, w, b)
    return torch.stack([dloss_dw.sum(), dloss_db.sum()])

def training_loop(n_epochs, optimizer, params, train_t_u, val_t_u, train_t_c, val_t_c):
    for epoch in range(1, n_epochs + 1):
        train_t_p = model(train_t_u, *params)
        train_loss = loss_fn(train_t_p, train_t_c)

        with torch.no_grad():
            val_t_p = model(val_t_u, *params)
            val_loss = loss_fn(val_t_p, val_t_c)
            assert val_loss.requires_grad == False

        optimizer.zero_grad()
        train_loss.backward()
        optimizer.step()
        training_stat = f'Train: {train_loss:.3f}  Validation: {val_loss:.3f}'
        print('{: <{padding}}'.format(training_stat, padding=75), end='\r', flush=True)

def calc_forward(t_u, t_c, is_train):
    with torch.set_grad_enabled(is_train):
        t_p = model(t_u, *params)
        loss = loss_fn(t_p, t_c)
    return loss
```

```python
# Trainning on new model

w = torch.ones(())
b = torch.zeros(())
w1 = torch.ones(())

params = torch.tensor([1.0, 1.0, 0.0], requires_grad=True)
learning_rate = 1e-1
optimizer = torch.optim.Adam([params], lr=learning_rate)

training_loop(
    n_epochs = 3000,
    optimizer = optimizer,
    params = params,
    train_t_u = train_t_un,
    val_t_u = val_t_un,
    train_t_c = train_t_c,
    val_t_c = val_t_c)

t_un = 0.1 * t_u
t_p = model(t_un, *params)
plot_result(t_p, t_u, t_c)
```
